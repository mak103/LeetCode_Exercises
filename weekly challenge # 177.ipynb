5169. Number of Days Between Two Dates (easy)

Write a program to count the number of days between two dates.

The two dates are given as strings, their format is YYYY-MM-DD as shown in the examples.

 

Example 1:

Input: date1 = "2019-06-29", date2 = "2019-06-30"
Output: 1
Example 2:

Input: date1 = "2020-01-15", date2 = "2019-12-31"
Output: 15
 

Constraints:

The given dates are valid dates between the years 1971 and 2100.

```python
# AC
class Solution(object):
    def daysBetweenDates(self, date1, date2):
        """
        :type date1: str
        :type date2: str
        :rtype: int
        """
        year1 = [0,31,28,31,30,31,30,31,31,30,31,30,31]
        year2 = [0,31,29,31,30,31,30,31,31,30,31,30,31]
        dict_years = {}
        for i in range(1971,2101):
            if i % 100 != 0:
                if i % 4 == 0:
                    dict_years[i] = year2
                else:
                    dict_years[i] = year1
            else:
                if i % 400 == 0:
                    dict_years[i] = year2
                else:
                    dict_years[i] = year1
        y1, m1, d1 = int(date1[:4]), int(date1[5:7]), int(date1[8:10])
        y2, m2, d2 = int(date2[:4]), int(date2[5:7]), int(date2[8:10])
        if y1 > y2:
            a, b, c = y1, m1, d1
            y1, m1, d1 = y2, m2, d2
            y2, m2, d2 = a, b, c
        elif y1 == y2 and m1 > m2:
            a, b, c = y1, m1, d1
            y1, m1, d1 = y2, m2, d2
            y2, m2, d2 = a, b, c
        elif y1 == y2 and m1==m2 and d1 > d2:
            a, b, c = y1, m1, d1
            y1, m1, d1 = y2, m2, d2
            y2, m2, d2 = a, b, c
        res = 0
        for i in range(y1, y2):
            res += sum(dict_years[i])
        for i in range(1, m2):
            res += dict_years[y2][i]
        res += d2
        for i in range(1, m1):
            res -= dict_years[y1][i]
        res -= d1
        return res
```

5172. Largest Multiple of Three (hard)

Given an integer array of digits, return the largest multiple of three that can be formed by concatenating some of the given digits in any order.

Since the answer may not fit in an integer data type, return the answer as a string.

If there is no answer return an empty string.

 

Example 1:

Input: digits = [8,1,9]
Output: "981"
Example 2:

Input: digits = [8,6,7,1,0]
Output: "8760"
Example 3:

Input: digits = [1]
Output: ""
Example 4:

Input: digits = [0,0,0,0,0,0]
Output: "0"
 

Constraints:

1 <= digits.length <= 10^4
0 <= digits[i] <= 9
The returning answer must not contain unnecessary leading zeros.

```python
# AC
class Solution(object):
    def find_nums(self, digits):
        if sum(digits) % 3 == 0:
            return digits
        if len(digits) <= 1:
            return []
        i = 0
        while i < len(digits):
            if digits[i] % 3 == 0:
                i += 1
                continue
            if sum(digits[:i]+digits[i+1:]) % 3 == 0:
                return digits[:i]+digits[i+1:]
            while i+1<len(digits) and digits[i] == digits[i+1]:
                i += 1
            i += 1
        i = 0
        while i < len(digits):
            if digits[i] % 3 == 0:
                i += 1
                continue
            d = self.find_nums(digits[:i]+digits[i+1:])
            if len(d) > 0:
                return d
            while i+1<len(digits) and digits[i] == digits[i+1]:
                i += 1
            i += 1
        return []
    
    def largestMultipleOfThree(self, digits):
        """
        :type digits: List[int]
        :rtype: str
        """
        digits = sorted(digits)
        n_0 = 0
        while len(digits) > 0 and digits[0] == 0:
            digits.pop(0)
            n_0 += 1
        nums = self.find_nums(digits)
        if len(nums) == 0 and n_0 == 0:
            return ""
        else:
            nums = nums[::-1]
            res = 0
            for n in nums:
                res = res *10 + n
            res = res * 10**n_0
            if res == 0:
                return "0"
            return str(res)
```

5171. Closest Divisors (medium)

Given an integer num, find the closest two integers in absolute difference whose product equals num + 1 or num + 2.

Return the two integers in any order.

 

Example 1:

Input: num = 8
Output: [3,3]
Explanation: For num + 1 = 9, the closest divisors are 3 & 3, for num + 2 = 10, the closest divisors are 2 & 5, hence 3 & 3 is chosen.
Example 2:

Input: num = 123
Output: [5,25]
Example 3:

Input: num = 999
Output: [40,25]
 

Constraints:

1 <= num <= 10^9

```python
# Excessed time limite
class Solution(object):
    def check(self, num1):
        s1 = num1 ** 0.5
        if s1 % 1 == 0:
            return [s1, s1]
        s1 = int(s1)
        #print(num1, s1)
        a, b = s1,s1
        while a*b != num1:
            print(a,b, a*b)
            if a*b > num1:
                a -= 1
            else:
                b += 1
        return a, b
    
    def closestDivisors(self, num):
        """
        :type num: int
        :rtype: List[int]
        """
        a1, b1 = self.check(num+1)
        a2, b2 = self.check(num+2)
        if abs(a1-b1) <= abs(a2-b2):
            return [int(a1), int(b1)]
        else:
            return [int(a2), int(b2)]
```
