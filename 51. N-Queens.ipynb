The n-queens puzzle is the problem of placing n queens on an n×n chessboard such that no two queens attack each other.



Given an integer n, return all distinct solutions to the n-queens puzzle.

Each solution contains a distinct board configuration of the n-queens' placement, where 'Q' and '.' both indicate a queen and an empty space respectively.

Example:

Input: 4
Output: [
 [".Q..",  // Solution 1
  "...Q",
  "Q...",
  "..Q."],

 ["..Q.",  // Solution 2
  "Q...",
  "...Q",
  ".Q.."]
]
Explanation: There exist two distinct solutions to the 4-queens puzzle as shown above.

```python
# My first solution (248ms, 9.91%, 12.9MB, 5.48%)

class Solution(object):
    def fillBoard(self, possible):
        res1 = []
        for i in range(self.n):
            temp = ['.']*self.n
            temp[possible[i]] = 'Q'
            res1.append("".join(temp))
        self.res.append(res1)
        return

    def check(self, possible, dict_):
        import copy
        i = len(possible)
        if i == self.n:
            self.fillBoard(possible)
            return
        filled = dict_[i]
        filled = list(set(filled))
        del dict_[i]
        for j in range(self.n):
            if j in filled:
                continue
            else:
                dict_1 = copy.deepcopy(dict_)
                for k in range(i+1, self.n):
                    dict_1[k].append(j)
                    if j-(k-i) >= 0:
                        dict_1[k].append(j-(k-i))
                    if j+(k-i) < self.n:
                        dict_1[k].append(j+(k-i))
                possible.append(j)
                self.check(possible, dict_1.copy())
                possible.pop()
        return 


    def solveNQueens(self, n):
        """
        :type n: int
        :rtype: List[List[str]]
        """
        import copy
        self.n = n
        if n == 0:
            return []
        if n == 1:
            return [['Q']]
        self.res = []
        dict_ = {}
        for i in range(1, n):
            dict_[i] = []
        
        for i in range(n):
            possible = [i]
            dict_1 = copy.deepcopy(dict_)
            for j in range(1, n):
                dict_1[j].append(i)
                if i-j >= 0:
                    dict_1[j].append(i-j)
                if i+j < n:
                    dict_1[j].append(i+j)
            self.check(possible, dict_1.copy())
        return self.res
```

```python
# My second solution (148ms, 17.30%, 12.9MB, 5.48%)
# 利用对称性

class Solution(object):
    def fillBoard(self, possible):
        res1, res2 = [], []
        for i in range(self.n):
            temp = ['.']*self.n
            temp[possible[i]] = 'Q'
            res1.append("".join(temp))
            res2.append("".join(temp[::-1]))
        if self.n%2 != 0 and possible[0] == int(self.n/2):
            self.res.insert(self.mark, res1)
            self.mark += 1
        else:
            self.res.insert(self.mark, res2)
            self.res.insert(self.mark, res1)
            self.mark += 1
        return

    def check(self, possible, dict_):
        import copy
        i = len(possible)
        if i == self.n:
            self.fillBoard(possible)
            return
        filled = dict_[i]
        filled = list(set(filled))
        del dict_[i]
        for j in range(self.n):
            if j in filled:
                continue
            else:
                dict_1 = copy.deepcopy(dict_)
                for k in range(i+1, self.n):
                    dict_1[k].append(j)
                    if j-(k-i) >= 0:
                        dict_1[k].append(j-(k-i))
                    if j+(k-i) < self.n:
                        dict_1[k].append(j+(k-i))
                possible.append(j)
                self.check(possible, dict_1.copy())
                possible.pop()
        return 


    def solveNQueens(self, n):
        """
        :type n: int
        :rtype: List[List[str]]
        """
        import copy
        self.mark = 0
        self.n = n
        if n == 0:
            return []
        if n == 1:
            return [['Q']]
        self.res = []
        dict_ = {}
        for i in range(1, n):
            dict_[i] = []
        if n%2 == 0:
            m = n/2
        else:
            m = int(n/2)+1

        for i in range(m):
            possible = [i]
            dict_1 = copy.deepcopy(dict_)
            for j in range(1, n):
                dict_1[j].append(i)
                if i-j >= 0:
                    dict_1[j].append(i-j)
                if i+j < n:
                    dict_1[j].append(i+j)
            self.check(possible, dict_1.copy())
        return self.res
```
